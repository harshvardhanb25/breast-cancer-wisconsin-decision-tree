### Implmenting Decision Tree Training Algorithm

```python
class Tree:
    '''
    Class for decision tree classifier.
    '''
    class Node:
        '''
        Class for decision tree node.
        '''
        def __init__(self, feature, threshold, left, right):
            '''
            feature: feature to split on
            threshold: threshold to split on
            left: left subtree
            right: right subtree
            '''
            self.feature = feature
            self.threshold = threshold
            self.left = left
            self.right = right

        def __str__(self):
            return f'Tree.Node(feature={self.feature}, threshold={self.threshold}, \n\t , left={self.left},\n\t ,right={self.right}))'
        

    class Leaf:
        def __init__(self, label):
            '''
            label: label of the leaf node
            '''
            self.label = label

        def __str__(self):
            return str(self.label)

    def __init__(self, data, label_col, max_depth, criterion):
        '''
        data: pandas dataframe
        label_col: label column name
        max_depth: maximum depth of the tree
        criterion: splitting criterion
        '''
        self.data = data
        self.label_col = label_col
        self.max_depth = max_depth
        self.criterion = criterion
        self.root = self.fit()

        
    def majority_class(self, data):
        '''
        data: pandas dataframe
        return: majority class in the data
        '''
        return data[self.label_col].mode()[0]
    
    def feature_uncertainty_score(self, data):
        '''
        data: pandas dataframe
        return: uncertainty score of the data
        '''
        label_counts = data[self.label_col].value_counts()
        if self.criterion == 'gini':
            # gini impurity
            score = 1
            for label in label_counts:
                score -= (label/len(data))**2
            return score
        elif self.criterion == 'entropy':
            # entropy
            score = 0
            for label in label_counts:
                score -= (label/len(data)) * np.log2(label/len(data))
            return score
        elif self.criterion == 'misclassification':
            # misclassification error
            score = 1
            for label in label_counts:
                score = min(score, label/len(data))
            return score
        else:
            raise ValueError('Invalid criterion')
        
    def gain(self, data, feature, threshold):
        '''
        data: pandas dataframe
        feature: feature to split on
        threshold: threshold to split on
        return: information gain
        '''
        uncertainty = self.feature_uncertainty_score(data)
        left = data[data[feature] <= threshold]
        right = data[data[feature] > threshold]
        left_score = self.feature_uncertainty_score(left)
        right_score = self.feature_uncertainty_score(right)
        return uncertainty - (len(left)/len(data))*left_score - (len(right)/len(data))*right_score

    def best_split(self, data):
        '''
        data: pandas dataframe
        return: best feature and threshold to split on
        '''
        best_gain = -np.inf
        best_feature = None
        best_threshold = None
        for feature in data.columns:
            if feature != self.label_col:
                for threshold in data[feature]:
                    gain = self.gain(data, feature, threshold)
                    if gain > best_gain:
                        best_gain = gain
                        best_feature = feature
                        best_threshold = threshold
        return best_feature, best_threshold
    
    def fit(self):
        '''
        return: root node of the decision tree
        '''
        if self.max_depth == 0 or len(self.data) == 0:
            return self.Leaf(self.majority_class(self.data))
        feature, threshold = self.best_split(self.data)
        # print(feature, threshold)
        left = self.data[self.data[feature] <= threshold]
        right = self.data[self.data[feature] > threshold]
        left = left.drop([feature], axis=1)
        right = right.drop([feature], axis=1)

        if len(left) == 0 or len(right) == 0:
            return self.Leaf(self.majority_class(self.data))
        else:
            left = Tree(left, self.label_col, self.max_depth-1, self.criterion).root
            right = Tree(right, self.label_col, self.max_depth-1, self.criterion).root
            return self.Node(feature, threshold, left, right)
        
    def predict(self, data):
        '''
        data: pandas dataframe
        return: list of predictions
        '''
        predictions = []
        for index, row in data.iterrows():
            node = self.root
            while isinstance(node, self.Node):
                if row[node.feature] <= node.threshold:
                    node = node.left
                else:
                    node = node.right
            predictions.append(node.label)

        return predictions
    
    def __str__(self):
        return str(self.root)
    
```

### Spliting the dataset into training and testing data

```python
# train test split 90% train 10% test
from sklearn.model_selection import train_test_split
data_train, data_test = train_test_split(breast_cancer_df, test_size=0.1, random_state=42)
```

### Training the Decision Tree Classifier

```python
def best_depth(label_col, criterion, folds, max_depth=10):
    '''
    label_col: label column name
    criterion: splitting criterion
    folds: list of folds
    max_depth: maximum depth of the tree
    return: best depth and best accuracy
    '''
    best_depth = 0
    best_accuracy = 0
    
    for depth in range(max_depth):
        accuracy = 0
        
        for i in range(len(folds)):
            data_val = folds[i]
            data_train = pd.concat(folds[:i] + folds[i+1:])
            tree = Tree(data_train, label_col, depth, criterion)
            predictions = tree.predict(data_val.drop([label_col], axis=1))
            actual = data_val[label_col].tolist()
            accuracy += sum([1 for i in range(len(predictions)) if predictions[i] == actual[i]])/len(predictions)
        accuracy /= len(folds)
        print('Depth:', depth, 'Accuracy:', accuracy)
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_depth = depth
    return best_depth, best_accuracy

def best_criterion_depth(data: pd.DataFrame, label_col, k=5, max_depth=10):
    '''
    data: pandas dataframe
    label_col: label column name
    k: number of folds
    max_depth: maximum depth of the tree
    return: best criterion and best depth
    '''
    folds = []
    fold_size = len(data)//k
    for i in range(k):
        folds.append(data.iloc[i*fold_size:(i+1)*fold_size])
    best_criterion, best_criterion_best_depth = None, 0
    best_accuracy = 0

    for criterion in ['gini', 'entropy', 'misclassification']:
        print('Criterion:', criterion)
        criterion_best_depth, criterion_best_accuracy = best_depth(label_col, criterion, folds, max_depth)
        print('best depth:', criterion_best_depth, 'best accuracy:', criterion_best_accuracy)
        if criterion_best_accuracy > best_accuracy:
            best_accuracy = criterion_best_accuracy
            best_criterion = criterion
            best_criterion_best_depth = criterion_best_depth
    return best_criterion, best_criterion_best_depth
       
```
